# Linear Algebra: Vectors



```python
import numpy as np
```

```python
def l1_norm(x):
    x_norm = np.abs(x)
    x_norm = np.sum(x_norm)
    
    return x_norm
```

```python
def l2_norm(x):
    x_norm = np.sum(x**2)
    x_norm = np.sqrt(x_norm)
    
    return x_norm
```

```python
def angle(x, y):
    v = np.inner(x, y) / (l2_norm(x) * l2_norm(y))
    theta = np.arccos(v)
    
    return theta
```
