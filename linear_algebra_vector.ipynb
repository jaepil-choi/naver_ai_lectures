# Linear Algebra: Vectors



```python
import numpy as np
```

```python
def l1_norm(x):
    x_norm = np.abs(x)
    x_norm = np.sum(x_norm)
    
    return x_norm
```

```python
def l2_norm(x):
    x_norm = np.sum(x**2)
    x_norm = np.sqrt(x_norm)
    
    return x_norm
```

```python
def angle(x, y):
    v = np.inner(x, y) / (l2_norm(x) * l2_norm(y))
    theta = np.arccos(v)
    
    return theta
```

## 행렬의 내적 inner product 

`matmul` 쓰면 되는데, `@`도 된다. 

```python
a = np.random.choice(10, size=(3, 4, 5))
a
```

```python
b = np.random.choice(10, size=(3, 5, 4))
b
```

```python
np.matmul(a, b)
```

```python
a @ b
```

`matmul`과 `@`는 같은 것을 볼 수 있다. 

아래의 `dot`은 다르다. 

```python
np.dot(a, b)
```

여기선 차원을 이미 transposed 된 것으로 맞춰준 다음 계산을 하였다.

그래서 위에서 한 것 처럼 transpose 시킨 상태에선 되려 에러가 난다.

```python
b = b.reshape(3, 4, 5)

np.inner(a, b)
```

### 유사역행렬 pseudo inverse matrix 

Moore-Penrose 무어 펜로즈 역행렬 (A+ 라고 표기)

행과 열의 숫자가 다를 경우에도 차원을 되돌리는 개념을 사용 가능. 



행의 갯수가 더 많은 경우, 열의 갯수가 더 많은 경우에 따라 계산 방식의 순서가 달라진다. 



#### 질문 

np.linalg.inv(A.T @ A) 는 항상 존재하는가? 



```python
B = np.random.choice(10, size=(3, 4))
B
```

```python
B_pinv = np.linalg.pinv(B) # 유사역행렬
B_pinv
```

```python
B @ B_pinv # 항등행렬이 나온다. 
```

```python
B_pinv @ B # 항등행렬이 안나온다. 
```

위에서 확인할 수 있는 것 같이 

유사역행렬을 계산할 때 n > m인지 n < m 인지에 따라 순서가 달라진다. 

### 응용 1 : 연립방정식 풀기 

변수의 갯수가 식의 갯수보다 많을 경우 부정, 해가 무수히 많다고 배웠을텐데, 

유사역행렬을 이용하여 그 부정의 해 중 하나를 구할 수 있다. 

!

식의 갯수보다 (row, `n`) 변수의 갯수 (column, `m`)이 더 많은 상황이다. `n <= m`

따라서 `A @ A+ = I` 가 된다.
