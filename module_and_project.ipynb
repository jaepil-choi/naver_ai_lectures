## Module 

그냥 .py 파일을 의미한다. `import`로 호출하면 된다. 

`import`로 모든 코드를 메모리에 로드하고 함수나 클래스를 부르는 것. 

참고로, 파이썬 실행을 시키면 .pyc 파일이 만들어진다. 이 pycache 파일은 back단에서 더 쉽게 실행하도록 기계어로 번역해주는 것이다. 

모든 코드를 로드하지 않고 필요한 부분만 호출하기 위해 

```python
from foo import bar
```

와 같이 해준다. 

good readability를 위해선 `*`로 모두 호출하거나 별도 모듈을 호출해 그것만 달랑 적어주는 것 보단 module을 불러 어떤 모듈의 어떤 함수인지 확실히 밝혀주는 것이 좋다. 

## Package

다양한 모듈의 합. 폴더로 연결 됨. 

`__init__.py`, `__main__.py`와 같은 키워드 파일명이 사용된다. 



- `__init__.py`
    - 현재 폴더가 패키지임을 알리는 초기화 스크립트
    - 3.3 이후부턴 `__init__.py`가 없어도 패키지로 간주한다. 이전엔 필수였다.  

원래 만든다면, 내부는 아래와 같이 어떤 모듈을 사용할 것인지를 밝혀주면 되었었다. 

```python
__all__ = ['image', 'sound', 'stage']

from . import image
from . import sound
from . import stage
```

- `__main__.py`
    - 이걸 넣어두면 터미널에서 실행시킬 때 폴더명으로도 실행이 된다. 
    
아래와 같이 작성한다. 

```python
from sound import echo

if __name__ == '__main__':
    print('hello world')
    echo.play_echo()
```

`import`시 상대참조/절대참조 둘 다 가능하다. 

## 오픈소스 라이브러리 사용하기 

가상환경을 만들자. 

필요한 패키지만 설치하는 환경. 다양한 패키지 관리 도구를 사용할 수 있음. 

가장 대표적으론 `virtualenv` + `pip` 조합이 있다. 레퍼런스와 패키지 개수에서 압도적이다. 

상용 가상환경 도구로는 `conda`가 있다. windows에서 용이하다. 왜냐면 pip에는 가끔 컴파일된 코드가 안들어가 있을 때가 있는데 아나콘다에선 컴파일된 도구를 같이 가지도록 만들어져있다.
