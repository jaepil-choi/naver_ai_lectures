# 딥러닝 학습방법 이해하기 

신경망은 기본적으로 비선형. 

하지만 분해해보면 선형모델과 비선형 모델이 섞여있다. 

어떻게 신경망은 비선형 패턴을 모델화 하는 것일까? 

바로 선형모델에 활성화함수 (activation function)을 결합함으로써 비선형을 표현할 수 있게 되는 것이다. 

## 순전파 수식으로 분해

O(output) = X * W + b 선형모델을 기본으로 한다. 



차원이 n * d의 데이터에서 n * p롤 바뀐다. 

W, 가중치 행렬이 바로 d개의 column을 가진 X를 p개의 output으로 연결시켜주는 화살표들이다. 화살표가 곧 가중치이다. 



activation function으로 softmax 함수를 쓰면 모델의 출력을 확률로 해석할 수 있게 된다. 

이를 통해 분류 모델을 풀 수 있다. 

이 부분은 강의에서 자세히 다뤄지지 않고 있다. 밑바닥 딥러닝을 참고하자. 

다루는 내용들: 
- softmax를 적용할 때 max값을 빼줘서 overflow를 막는 내용
- 추론 할 때는 (신경망의 맨 끝단) softmax가 굳이 필요없고 one-hot 으로 결과만 내면 된다는 부분
- 선형 모델로 나온 결과물을 활성화 함수를 씌우면 on/off가 되며 perceptron이 된다. 이걸로 비선형을 나타낸다. 
    - 활성화 함수는 실수공간 위에 정의된 **비선형** 함수로서 딥러닝에서 매우 중요한 개념이다. 
    - 활성화 함수를 쓰지 않으면 딥러닝은 선형모형과 차이가 전혀 차이가 없다. 
    - sigmoid나 tanh보다 딥러닝에선 ReLU가 많이 쓰인다. 
- 딥러닝, 즉 multi layer 는 선형모델 결과 Z에 활성화 함수를 씌워 나온 결과를 모은 잠재벡터 H를 또 선형변환하고 또 선형변환하고 해서 multi-layer perceptron 딥러닝이다. 



앞에서 말한 것 처럼, 선형 모델을 계속 쓰는데 그 중간에 계속 비선형인 활성화 함수를 써줘야 한다. 

이렇게 앞으로 보내는 것을 순전파라고 부른다. forward propagation. 학습이 아니고 주어진 입력이 왔을 때 출력하는 과정이다. 

### 왜 여러 층을 쌓는가? 

이론적으로는 2층 신경망으로도 임의의 연속함수를 근사할 수 있다. (universal approximation theorem이라 부른다.)

칭이 깊을수록 목적함수를 근사하는데 필요한 뉴런의 숫자가 훨씬 빨리 줄어들어 좀 더 효율적으로 학습이 가능. 

층이 얕으면 필요한 뉴런의 숫자가 기하급수적으로 늘어나 wide한 신경망이 되어야 한다. 

## 역전파 수식으로 분해

aL/aW (a는 편미분 표시) 를 할 때 역전파가 사용된다. 

역전파는 chain rule을 통해 gradient vector를 전달한다.

연쇄법칙 기반 자동미분을 사용한다. chain-rule based auto-differentiation 


참고: 각 뉴런의 값을 tensor라 부르는데, 각 tensor의 값은 메모리에 저장되어야 back propagation이 작동한다. 그래서 back propagation은 forward propagation보다 메모리를 더 사용한다.
